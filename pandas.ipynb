*************merge***************

```python
var1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var1
```

```python
var1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})#atleast one part common id
var2=pd.DataFrame({"A":[1,2,3,4,5],"C":[21,22,33,24,25]})
pd.merge(var1,var2,on="A")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})#atleast one part common id
var2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,33,25]})#only common part is done
pd.merge(var1,var2,on="A")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})#atleast one part common id
var2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,33,25]})#only common part is done
pd.merge(var1,var2,how="outer")#left,right,inner,outer
```

```python
#tells which data is present
pd.merge(var1,var2,how="outer",indicator=True)
```

```python
pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","python"))
```
